```python
#Análise em base de dados da Chegada de Turistas Internacionais

#importação de bibliotecas que trabalham com dados tabulados
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importação de datasets de 2010 até 2019 
dados10 = pd.read_csv('chegadas_2010.csv', delimiter=';')
dados11 = pd.read_csv('chegadas_2011.csv', delimiter=';')
dados12 = pd.read_csv('chegadas_2012.csv', delimiter=';')
dados13 = pd.read_csv('chegadas_2013.csv', delimiter=';')
dados14 = pd.read_csv('chegadas_2014.csv', delimiter=';')
dados15 = pd.read_csv('chegadas_2015.csv', delimiter=';')
dados16 = pd.read_csv('chegadas_2016.csv', delimiter=';')
dados17 = pd.read_csv('chegadas_2017.csv', delimiter=';')
dados18 = pd.read_csv('chegadas_2018.csv', delimiter=';')
dados19 = pd.read_csv('chegadas_2019.csv', delimiter=';')
```

```python
#relação da troca de nomes das colunas
#passo necessário por as colunas terem nomes diferentes 
relacao_colunas = {'cod continente' : 'Ordem continente', 'cod pais' : 'Ordem país', 'cod via' : 'Ordem via de acesso', 'ano' : 'Ano', 'cod mes' : 'Ordem mês', 'cod uf' : 'Ordem UF', 'Via' : 'Via de acesso'}

#concatenação dos datasets
dados = pd.concat([dados10.rename(columns=relacao_colunas), dados11.rename(columns=relacao_colunas), dados12.rename(columns=relacao_colunas), dados13.rename(columns=relacao_colunas), dados14.rename(columns=relacao_colunas), dados15.rename(columns=relacao_colunas), dados16.rename(columns=relacao_colunas), dados17.rename(columns=relacao_colunas), dados18.rename(columns=relacao_colunas), dados19.rename(columns=relacao_colunas)])

#o head imprime as 5 primeiras linhas do dataSet
dados.head()
```

```python
#remove a coluna Unnamed
dados = dados.drop(columns=['Unnamed: 12'])
dados.head()
```

```python
#normaliza os dados que representam o mesmo valor
dados['Mês'] = dados['Mês'].replace(to_replace=['janeiro'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['fevereiro'],value='Fevereiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['março'],value='Março')
dados['Mês'] = dados['Mês'].replace(to_replace=['abril'],value='Abril')
dados['Mês'] = dados['Mês'].replace(to_replace=['maio'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['junho'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['julho'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['agosto'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['setembro'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['outubro'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['novembro'],value='Janeiro')
dados['Mês'] = dados['Mês'].replace(to_replace=['dezembro'],value='Janeiro')
dados.head()
```

```python
#retorna todas as diferentes valores prenchidos nas colunas
dados['Continente'].unique()
dados['Ordem continente'].unique()
dados['País'].unique()
dados['UF'].unique()
dados['Ordem UF'].unique()
dados['Via de acesso'].unique()
dados['Ordem via de acesso'].unique()
dados['Ano'].unique()
dados['Mês'].unique()
dados['Ordem mês'].unique()
dados['Chegadas'].unique()
```
